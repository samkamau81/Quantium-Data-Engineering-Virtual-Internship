## Customer Category Vs Total Sales

```python
print(df.groupby(['PREMIUM_CUSTOMER'])['PREMIUM_CUSTOMER'].count())
```

```python
len(df)

Budget=0
Mainstream=0
Premium=0

for i in range(len(df)):
    x = df.iloc[i,11] 
      
    if x == 'Budget' :
        Budget_tsale=df.iloc[i,7]
        Budget=Budget+Budget_tsale
        
        
    elif x == 'Mainstream' :
        Mainstream_tsale=df.iloc[i,7]
        Mainstream=Mainstream+Mainstream_tsale    

            
    elif x == 'Premium' :
        Premium_tsale=df.iloc[i,7]
        Premium=Premium+Premium_tsale
        
print ('Budget =',Budget)
print ('Mainstream =',Mainstream)
print ('Premium =',Premium)
```

```python
fig= plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.set_xlabel('Customer Category')
ax.set_ylabel('Total Sales')
ax.set_title('Graph 2: Total Sales Vs Customer Category')
CUSTOMER = ['Budget','Mainstream','Premium']
SALES = [Budget, Mainstream, Premium]
ax.bar(CUSTOMER, SALES)

plt.show()

labels = ['Premium', 'Mainstream', 'Budget']
sizes = [Premium, Mainstream, Budget]
colors = ['yellowgreen', 'gold', 'lightskyblue']
patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90)
plt.legend(patches, labels, loc="best")
plt.axis('equal')
plt.tight_layout()
plt.show()
```
