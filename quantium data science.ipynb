```python
#Graphs

#Graph 1: Product Quality vs Total Sales

x = result.TOT_SALES
y = result.PROD_QTY

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y, color='lightblue', linewidth=3)

ax.scatter([2,4,6],[5,15,25],color='darkgreen',marker='^')

ax.set_xlim(1, 6.5)
ax.set_xlabel('Total Sales')
ax.set_ylabel('Product Quality')
ax.set_title('Product Quality Vs Total Sales')
plt.savefig('Graph.png')

plt.show()
```

```python
# Graph 2:The number of clients according to choices
print(result.groupby(['PREMIUM_CUSTOMER'])['PREMIUM_CUSTOMER'].count())
Premium =result['PREMIUM_CUSTOMER'].str.contains('Premium').value_counts()[True] #18922
Mainstream = result['PREMIUM_CUSTOMER'].str.contains('Mainstream').value_counts()[True] #29245
Budget = result['PREMIUM_CUSTOMER'].str.contains('Budget').value_counts()[True] #18922

labels = ['Premium', 'Mainstream', 'Budget']
sizes = [Premium, Mainstream, Budget]
colors = ['yellowgreen', 'gold', 'lightskyblue']
patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90)
plt.legend(patches, labels, loc="best")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Graph 3:Lifestage Group count graphs
print(result.groupby(['LIFESTAGE'])['LIFESTAGE'].count())
NEW_FAMILIES =result['LIFESTAGE'].str.contains('NEW FAMILIES').value_counts()[True] 
YOUNG_SINGLES_COUPLES = result['LIFESTAGE'].str.contains('YOUNG SINGLES/COUPLES').value_counts()[True] 
RETIREES = result['LIFESTAGE'].str.contains('RETIREES').value_counts()[True]
OLDER_SINGLES_COUPLES = result['LIFESTAGE'].str.contains('OLDER SINGLES/COUPLES').value_counts()[True] 
MIDAGE_SINGLES_COUPLES = result['LIFESTAGE'].str.contains('MIDAGE SINGLES/COUPLES').value_counts()[True] 

labels = ['NEW FAMILIES', 'YOUNG SINGLES/COUPLES', 'RETIREES','OLDER SINGLES/COUPLES','MIDAGE SINGLES/COUPLES']
sizes = [NEW_FAMILIES,YOUNG_SINGLES_COUPLES, RETIREES,OLDER_SINGLES_COUPLES, MIDAGE_SINGLES_COUPLES]
colors = ['yellowgreen', 'gold', 'lightskyblue','green','purple']
patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90)
plt.legend(patches, labels, loc="best")
plt.axis('equal')
plt.tight_layout()
plt.show()
```
